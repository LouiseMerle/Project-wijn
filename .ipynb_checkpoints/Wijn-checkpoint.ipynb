```python
wijnen = pd.concat([red, white], ignore_index=True, sort=False)
```

```python
del red['type']
del white['type']
```

```python
y_red = red['score']

x_red = red[red.columns.difference(['quality', 'score'])]
```

```python
y_white = white['score']

x_white = white[white.columns.difference(['quality', 'score'])]
```

```python
'''
y_white = white_dummy[['quality_4', 'quality_5', 'quality_6','quality_7', 'quality_8', 'quality_9']]
x_white = white_dummy[white_dummy.columns.difference(['quality_4', 'quality_5', 'quality_6','quality_7', 'quality_8', 'quality_9'])]
'''
```

```python
y_wijnen = wijnen['type']
x_wijnen = wijnen[wijnen.columns.difference(['type'])]

wijnen.to_csv('wijnen.csv')
```

```python
#rood wit split
X_wijn_train, X_wijn_test, y_wijn_train, y_wijn_test = train_test_split(x_wijnen, 
                                                                        y_wijnen, 
                                                                        test_size = 0.20)
```

```python
# white quality split
X_white_train, X_white_test, y_white_train, y_white_test = train_test_split(x_white, 
                                                                        y_white, 
                                                                        test_size = 0.20)
```

```python
# 
X_red_train, X_red_test, y_red_train, y_red_test = train_test_split(x_red, 
                                                                        y_red, 
                                                                        test_size = 0.20)
```

# red / white prediction

logistic

```python
wijn_model = LogisticRegression().fit(X_wijn_train, y_wijn_train)
y_pred_wijn = wijn_model.predict(X_wijn_test)
print('confusion matrix is \n', confusion_matrix(y_wijn_test, y_pred_wijn))
print('accuracy_score', balanced_accuracy_score(y_wijn_test, y_pred_wijn))

y_pred_proba = wijn_model.predict_proba(X_wijn_test)[::,1]

fpr, tpr, _ = metrics.roc_curve(y_wijn_test, y_pred_proba)
plt.plot(fpr,tpr, color='red')
auc = metrics.roc_auc_score(y_wijn_test, y_pred_proba)
print('auc', auc)

#px.scatter(x=y_wijn_test, y=y_pred_wijn)
```

```python
# rood / wit met tree model

tree_wijn = RandomForestClassifier().fit(X_wijn_train, y_wijn_train)
y_pred_test_wijn_tree = tree_wijn.predict(X_wijn_test)
print('confusion matrix is \n', confusion_matrix(y_wijn_test, y_pred_test_wijn_tree))
print('balanced accuracy score', balanced_accuracy_score(y_wijn_test, y_pred_test_wijn_tree))


y_pred_proba_wijn_tree = tree_wijn.predict_proba(X_wijn_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_wijn_test,  y_pred_proba_wijn_tree)
auc = metrics.roc_auc_score(y_wijn_test, y_pred_proba_wijn_tree)
print('auc', auc)
plt.plot(fpr,tpr, color='red')

#plt.figure()
#tree.plot_tree(tree_wijn)
#plot_tree(tree_wijn, filled=True)
#plt.show()
```

```python
# rood / wit met SVM 

clf_wijn = svm.SVC(gamma='auto', probability=True)
clf_wijn.fit(X_wijn_train, y_wijn_train)
print('score', clf_wijn.score(X_wijn_test, y_wijn_test))

y_pred_proba_wijn_SVC = clf_wijn.predict_proba(X_wijn_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_wijn_test,  y_pred_proba_wijn_SVC)
auc = metrics.roc_auc_score(y_wijn_test, y_pred_proba_wijn_SVC)
print('auc', auc)
plt.plot(fpr,tpr, color='r')
```

```python
# rood / wit met KNN

knn_wijn = KNeighborsClassifier(n_neighbors=3)

knn_wijn.fit(X_wijn_train, y_wijn_train)
y_wijn_pred = knn_wijn.predict(X_wijn_test)

print('confusion_matrix \n', confusion_matrix(y_wijn_test, y_wijn_pred))
print('accuracy_score', balanced_accuracy_score(y_wijn_test, y_wijn_pred))

y_pred_proba_wijn_tree = tree_wijn.predict_proba(X_wijn_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_wijn_test,  y_pred_proba_wijn_tree)
auc = metrics.roc_auc_score(y_wijn_test, y_pred_proba_wijn_tree)
print('auc', auc)
plt.plot(fpr,tpr, color='r')
```
